```python
import pandas as pd
import numpy as np
import string

from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time
```

```python
df = pd.read_csv('Data/nba_draft_first_round.csv')

stats = ['points', 'rebounds', 'assists', 'steals', 'blocks']
for stat in stats:
    df[stat] = ''
df
```

```python
def remove_punctuations(text):
    for punctuation in ['.', "'"]:
        text = text.replace(punctuation, '')
    return text

df['Player'] = df['Player'].apply(remove_punctuations)
player_list = df['Player'].str.replace(' ', '-').str.lower().to_list()
player_list
```

```python
driver = webdriver.Chrome(executable_path='C:/Users/17178/Downloads/chromedriver_win32/chromedriver.exe')

for i, player in enumerate(player_list):
    driver.get('https://www.sports-reference.com/cbb/players/{}-1.html'.format(player))
    try:
        points = driver.find_element_by_xpath('/html/body/div[2]/div[5]/div[2]/div[2]/table/tfoot/tr/td[26]').text
        rebounds = driver.find_element_by_xpath('/html/body/div[2]/div[5]/div[2]/div[2]/table/tfoot/tr/td[20]').text
        assists = driver.find_element_by_xpath('/html/body/div[2]/div[5]/div[2]/div[2]/table/tfoot/tr/td[21]').text
        steals = driver.find_element_by_xpath('/html/body/div[2]/div[5]/div[2]/div[2]/table/tfoot/tr/td[22]').text
        blocks = driver.find_element_by_xpath('/html/body/div[2]/div[5]/div[2]/div[2]/table/tfoot/tr/td[23]').text
        for stat in stats:
            if stat == 'points':
                df.at[i, stat] = points
            elif stat == 'rebounds':
                df.at[i, stat] = rebounds
            elif stat == 'assists':
                df.at[i, stat] = assists
            elif stat == 'steals':
                df.at[i, stat] = steals
            else:
                df.at[i, stat] = blocks
    except:
        continue
    try:
        points = driver.find_element_by_xpath('/html/body/div[2]/div[5]/div[2]/div[2]/table/tfoot/tr[1]/td[26]').text
        rebounds = driver.find_element_by_xpath('/html/body/div[2]/div[5]/div[2]/div[2]/table/tfoot/tr[1]/td[20]').text
        assists = driver.find_element_by_xpath('/html/body/div[2]/div[5]/div[2]/div[2]/table/tfoot/tr[1]/td[21]').text
        steals = driver.find_element_by_xpath('/html/body/div[2]/div[5]/div[2]/div[2]/table/tfoot/tr[1]/td[22]').text
        blocks = driver.find_element_by_xpath('/html/body/div[2]/div[5]/div[2]/div[2]/table/tfoot/tr[1]/td[23]').text
        for stat in stats:
            if stat == 'points':
                df.at[i, stat] = points
            elif stat == 'rebounds':
                df.at[i, stat] = rebounds
            elif stat == 'assists':
                df.at[i, stat] = assists
            elif stat == 'steals':
                df.at[i, stat] = steals
            else:
                df.at[i, stat] = blocks
    except:
        for stat in stats:
            df.at[i, stat] = np.nan
        continue

df
```

```python
df.to_csv('C:/Users/17178/Documents/DS Projects/FanDuel/NBA Draft Project/Data/nba_draft_first_round_with_stats.csv')
```
